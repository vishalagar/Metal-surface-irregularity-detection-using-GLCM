```python
%matplotlib inline
import math
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
from skimage import measure
from skimage.color import rgb2gray
from skimage.util import img_as_ubyte
from skimage import io
from skimage.feature import graycomatrix,graycoprops
```

```python
imgList = os.listdir('NEU-DET/train/images/scratches/')
```

```python
imgPath = []
for path in imgList:
    pathImg = f'NEU-DET/train/images/scratches/{path}'
    imgPath.append(pathImg)
imgPath.sort()
```

```python
ax = []
ay = []
az = []
aw = []
for patch in imgPath:
    image_rgb = io.imread(patch)
    image = img_as_ubyte(rgb2gray(image_rgb))
    glcm = graycomatrix(image, distances=[4], angles=[0], levels=256)
    ax.append(graycoprops(glcm, 'dissimilarity')[0, 0])
    ay.append(graycoprops(glcm, 'correlation')[0, 0])
    az.append(graycoprops(glcm, 'contrast')[0, 0])
    aw.append(graycoprops(glcm, 'homogeneity')[0, 0])
```

```python
bx = []
by = []
bz = []
bw = []
for patch in imgPath:
    image_rgb = io.imread(patch)
    image = img_as_ubyte(rgb2gray(image_rgb))
    glcm = graycomatrix(image, distances=[4], angles=[0.785398], levels=256)
    bx.append(graycoprops(glcm, 'dissimilarity')[0, 0])
    by.append(graycoprops(glcm, 'correlation')[0, 0])
    bz.append(graycoprops(glcm, 'contrast')[0, 0])
    bw.append(graycoprops(glcm, 'homogeneity')[0, 0])
```

```python
cx = []
cy = []
cz = []
cw = []
for patch in imgPath:
    image_rgb = io.imread(patch)
    image = img_as_ubyte(rgb2gray(image_rgb))
    glcm = graycomatrix(image, distances=[4], angles=[1.0472], levels=256)
    cx.append(graycoprops(glcm, 'dissimilarity')[0, 0])
    cy.append(graycoprops(glcm, 'correlation')[0, 0])
    cz.append(graycoprops(glcm, 'contrast')[0, 0])
    cw.append(graycoprops(glcm, 'homogeneity')[0, 0])
```

```python
dx = []
dy = []
dz = []
dw = []
for patch in imgPath:
    image_rgb = io.imread(patch)
    image = img_as_ubyte(rgb2gray(image_rgb))
    glcm = graycomatrix(image, distances=[4], angles=[2.0944], levels=256)
    dx.append(graycoprops(glcm, 'dissimilarity')[0, 0])
    dy.append(graycoprops(glcm, 'correlation')[0, 0])
    dz.append(graycoprops(glcm, 'contrast')[0, 0])
    dw.append(graycoprops(glcm, 'homogeneity')[0, 0])
```

```python
ex = []
ey = []
ez = []
ew = []
for patch in imgPath:
    image_rgb = io.imread(patch)
    image = img_as_ubyte(rgb2gray(image_rgb))
    glcm = graycomatrix(image, distances=[4], angles=[3.14159], levels=256)
    ex.append(graycoprops(glcm, 'dissimilarity')[0, 0])
    ey.append(graycoprops(glcm, 'correlation')[0, 0])
    ez.append(graycoprops(glcm, 'contrast')[0, 0])
    ew.append(graycoprops(glcm, 'homogeneity')[0, 0])
```

```python
fx = []
fy = []
fz = []
fw = []

for patch in imgPath:
    image_rgb = io.imread(patch)
    image = img_as_ubyte(rgb2gray(image_rgb))
    glcm = graycomatrix(image, distances=[4], angles=[4.18879], levels=256)
    fx.append(graycoprops(glcm, 'dissimilarity')[0, 0])
    fy.append(graycoprops(glcm, 'correlation')[0, 0])
    fz.append(graycoprops(glcm, 'contrast')[0, 0])
    fw.append(graycoprops(glcm, 'homogeneity')[0, 0])
```

```python
def assign_class(image_name):
    if 'scratches' in image_name:
        return 1
    elif 'inclusion' in image_name:
        return 2
    elif 'pitted_surface' in image_name:
        return 3
    elif 'rolled-in_scale' in image_name:
        return 4
    elif 'crazing' in image_name:
        return 5
    elif 'patches' in image_name:
        return 6
    else:
        return None

# Apply the function to create the 'Class' column
final4['Class'] = final4['image_name'].apply(assign_class)

# Print the DataFrame to verify the addition of the 'Class' column
# print(final4)
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
df = final4
# Assuming the data is stored in a DataFrame named df
# You can load the data using: df = pd.read_csv('your_data.csv')

# Selecting the relevant columns for clustering
# X = df[['ax', 'ay', 'bx', 'by', 'cx', 'cy', 'dx', 'dy', 'ex', 'ey', 'fx', 'fy']]

# Define the number of clusters (k)
k = 6

# Initialize KMeans model
kmeans = KMeans(n_clusters=k, random_state=0)

# Fit the model to the data
kmeans.fit(X)

# Get the cluster labels for each data point
labels = kmeans.labels_

# Add cluster labels to the DataFrame
df['Cluster'] = labels

# Print the cluster assignments
print(df[['ax', 'ay', 'bx', 'by', 'cx', 'cy', 'dx', 'dy', 'ex', 'ey', 'fx', 'fy', 'Cluster']])

# Visualize the clusters (for 2D data)
# Note: Since your data has 12 dimensions, visualization would be challenging.
# You may want to use dimensionality reduction techniques for visualization in practice.
# Create a color map
cmap = plt.cm.get_cmap('viridis', k)  # replace 'k' with the number of your clusters

# Plot each cluster separately
for i in range(k):
    cluster_data = X[labels == i]
    plt.scatter(cluster_data['ax'], cluster_data['ay'], color=cmap(i), label=f'Cluster {i}')

# Add labels and title
plt.xlabel('ax')
plt.ylabel('ay')
plt.title('KMeans Clustering')

# Add a legend
plt.legend()

# Display the plot
plt.show()

from sklearn.metrics import silhouette_score

# Calculate Silhouette Score
silhouette = silhouette_score(X, labels)

print(f'Silhouette Score: {silhouette}')

# Calculate Sum of Squared Distances
ssd = kmeans.inertia_

print(f'Sum of Squared Distances: {ssd}')
```
